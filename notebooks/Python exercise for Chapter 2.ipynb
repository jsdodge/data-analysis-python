```python
# import numpy library and name it "np".
import numpy as np
# import matplotlib.pyplot and name it "plt"
import matplotlib.pyplot as plt


# initialize a pseudorandom number generator
np.random.seed(2)

mu = 10
sigma = 1
n = 2500 

# generate n = 2500 random number (mu, sigma)
data_0 = np.random.randn(n) * sigma + mu



    
plt.subplot(2, 1, 1)
plt.plot(data_0, np.arange(len(data_0)) / 1000, '-')

# set the x-limit
plt.xlim((6, 15))

plt.title('raw data')
plt.ylabel('Trial / 1000')

plt.subplot(2, 1, 2)

plt.hist(data_0, range=(0, mu * 1.5), bins=100)

# draw a vertical dashed line to show the expected value
plt.axvline(mu, linestyle='--', color='r')


# set the x-limit
plt.xlim((6, 15))

plt.xlabel('Variable')
plt.ylabel('Occurrence')


# adjust spacing between subplots to minimize the overlaps
plt.tight_layout()
```

```python
# in order to take the average of each 5 data point we reshape the array into a 500 x 5 2-d array
# then we take the average of each col

data_reshape_1 = data_0.reshape(500, 5)
data_1 = np.mean(data_reshape_1, axis=1)



plt.subplot(2, 1, 1)
plt.plot(data_1, np.arange(len(data_1)) / 100, '-')

# set the x-limit
plt.xlim((6, 15))

plt.title('raw data')
plt.ylabel('Trial / 100')

plt.subplot(2, 1, 2)

plt.hist(data_1, range=(0, mu * 1.5), bins=100)

# draw a vertical dashed line to show the expected value
plt.axvline(mu, linestyle='--', color='r')

# set the x-limit

plt.xlim((6, 15))

plt.xlabel('Variable')
plt.ylabel('Occurrence')
plt.title('Mean of every 5 samples')

# adjust spacing between subplots to minimize the overlaps
plt.tight_layout()
```
